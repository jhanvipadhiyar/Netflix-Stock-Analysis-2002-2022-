```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns

from datetime import datetime
```

```python
df=pd.read_csv("C:/Users/Jhanvi Padhiyar/projects/Netflix Stock Analysis/NFLX.csv")
```

```python
print(df.isnull().sum())
```

```python
sns.set(rc={"figure.figsize":(10,5)})
```

```python
df['Data']=pd.to_datetime(df['Date'])
df=df.set_index('Date')
df.head()
```

```python
sns.lineplot(x=df.index,y=df["Volume"],label='Volume')
plt.title("Volume of stock versus time")
plt.show()
```

```python
high=sns.lineplot(x=df.index,y=df["High"],label="High Price")
plt.title('High price of the stock')
open=sns.lineplot(x=df.index,y=df["Open"],label="Open stock")
plt.title('Open stock')
plt.show(high,open)
```

```python
df.plot(y=['High','Close','Open'],title='Netflix Stock Price')
plt.show()
```

```python
df.index = pd.to_datetime(df.index)
fig,(ax1,ax2,ax3)=plt.subplots(3,figsize=(15,10))
df.groupby(df.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day',ylabel='Volume')
df.groupby(df.index.month).mean().plot(y='Volume',ax=ax2,xlabel='Month',ylabel='Volume')
df.groupby(df.index.year).mean().plot(y='Volume',ax=ax3,xlabel='Year',ylabel='Volume')
```
